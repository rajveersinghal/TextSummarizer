```python
def convert_examples_to_features(example_batch):
  inptut_encodings = tokenizer(example_batch['dialogue'], max_length=1024, truncation=True)

  with tokenizer.as_target_tokenizer():
    target_encodings = tokenizer(example_batch['summary'], max_length=128, truncation=True)

  return {
      'input_ids': inptut_encodings['input_ids'],
      'attention_mask': inptut_encodings['attention_mask'],
      'labels': target_encodings['input_ids']
  }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched=True)
```

```python
dataset_samsum_pt["train"]
```

```python
# Training

from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir="pegasus-samsum",
    num_train_epochs=1,
    warmup_steps=500,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    weight_decay=0.01,
    logging_steps=10,
    eval_strategy="steps",
    eval_steps=500,
    save_steps=int(1e6),
    gradient_accumulation_steps=16,
    report_to="none"

)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

trainer = Trainer(
    model=model_pegasus,
    args=trainer_args,
    tokenizer=tokenizer,
    data_collator=seq2seq_data_collator,
    train_dataset=dataset_samsum_pt["test"],
    eval_dataset=dataset_samsum_pt["validation"]
)
```

```python
from tqdm import tqdm

def generate_batch_sized_chunks(list_of_elements, batch_size):
    """
    Splits a list into consecutive batches of a given size.

    Args:
        list_of_elements (list): The full list to split.
        batch_size (int): Number of items per batch.

    Yields:
        list: A batch of elements of size `batch_size` (last batch may be smaller).
    """
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]


def calculate_metric_on_test_ds(
    dataset,
    metric,
    model,
    tokenizer,
    batch_size=16,
    device="cuda",
    column_text="article",
    column_summary="highlights"
):
    """
    Evaluates a seq2seq summarization model on a dataset using a metric (e.g., ROUGE).

    This function:
      - Splits the dataset into batches
      - Tokenizes input text
      - Generates summaries using the model
      - Decodes summaries
      - Adds predictions + references to the metric
      - Computes and returns the final metric score

    Args:
        dataset (Dataset): HuggingFace dataset containing text + summaries.
        metric (Metric): HuggingFace evaluate metric object (e.g., evaluate.load("rouge")).
        model (PreTrainedModel): The seq2seq model to evaluate.
        tokenizer (PreTrainedTokenizer): Tokenizer for encoding/decoding.
        batch_size (int, optional): Batch size for evaluation. Defaults to 16.
        device (str, optional): Device to run inference on ("cuda" or "cpu"). Defaults to "cuda".
        column_text (str, optional): Column name containing input text. Defaults to "article".
        column_summary (str, optional): Column name containing reference summaries. Defaults to "highlights".

    Returns:
        dict: Computed metric scores (e.g., ROUGE-1, ROUGE-2, ROUGE-L).
    """

    # Split dataset into batches for efficient processing
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    # Iterate through batches with progress bar
    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches),
        total=len(article_batches),
        desc="Evaluating"
    ):
        # Tokenize input text
        inputs = tokenizer(
            article_batch,
            max_length=1024,
            truncation=True,
            padding="max_length",
            return_tensors="pt"
        )

        # Move tensors to device and generate summaries
        summaries = model.generate(
            input_ids=inputs["input_ids"].to(device),
            attention_mask=inputs["attention_mask"].to(device),
            length_penalty=0.8,
            num_beams=8,
            max_length=128
        )

        # Decode generated token IDs into text
        decoded_summaries = [
            tokenizer.decode(
                s,
                skip_special_tokens=True,
                clean_up_tokenization_spaces=True
            )
            for s in summaries
        ]

        # Add predictions + references to metric
        metric.add_batch(
            predictions=decoded_summaries,
            references=target_batch
        )

    # Compute final metric score
    score = metric.compute()
    return score
```

```python
import evaluate

rouge_metric = evaluate.load("rouge")
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
```

```python
score = calculate_metric_on_test_ds(
    dataset_samsum["test"][0:10],
    rouge_metric,
    model_pegasus,
    tokenizer,
    batch_size=2,
    column_text="dialogue",
    column_summary="summary"
)

rouge_dict = {rn: score[rn] for rn in rouge_names}

pd.DataFrame(rouge_dict, index=["pegasus"])
```
